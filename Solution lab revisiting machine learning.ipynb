>Check for null values in all the columns

```python
nulls = pd.DataFrame(data.isna().sum()*100/len(data), columns=['percentage'])
nulls.sort_values('percentage', ascending = False)
```

>Exclude the following variables by looking at the definitions. Create a new empty list called drop_list. We will append this list and then drop all the columns in this list later:
> - OSOURCE - symbol definitions not provided, too many categories
> - ZIP CODE - we are including state already 

```python
drop_list = ['OSOURCE','ZIP']
drop_list 
```

```python
#Drop the columns in drop_list and check the new number of columns in the datset
data_2=data.drop(columns=drop_list)
data_2.shape
```

>Identify columns that over 85% missing values

```python
#Create the same table as before with all nulls
nulls = pd.DataFrame(data.isna().sum()*100/len(data), columns=['percentage'])
#Create a table from the previous one where I filter the relevant values
filterednulls = nulls[nulls['percentage']>=85].sort_values('percentage', ascending = False)
filterednulls
```

```python
filterednulls.shape
#There is 25 features with more than 85% missing values or NaNs
```

```python
columns_to_delete = filterednulls.index.tolist()
columns_to_delete
```

```python
#Drop the columns in drop_list and check the new number of columns in the datset
data_3=data_2.drop(columns=columns_to_delete)
data_3.shape
```

>Reduce the number of categories in the column GENDER. The column should only have either "M" for males, "F" for females, and "other" for all the rest
> - Note that there are a few null values in the column. We will first replace those null values using the code below:

```python
data_3[['GENDER']].value_counts()
```

```python

data_3['GENDER'] = data_3['GENDER'].fillna('F')
```

```python
data_3[['GENDER']].value_counts()
```

```python
data_3[['GENDER']]= data_3[['GENDER']].apply(lambda x: 'other' if (x[0] != 'F' and x[0] != 'M') else x, axis=1)
```

```python
data_3[['GENDER']].value_counts()
```
